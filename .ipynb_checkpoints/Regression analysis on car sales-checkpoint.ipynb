# Predicting the price of a used car depending on its features

### Import relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

### Loading the data

```python
data = pd.read_csv('Car data.csv')
data.head()
```

## Data Preprocessing

```python
data.describe(include='all')
```

-  There are missing values.
-  312 unique values are present in Model which is hard to implement in regression (312 dummies)
-  A lot of information from Model can be obtained from Brand and Year so it can be dropped.

### Dropping columns with less/no dependency

```python
data = data.drop(['Model'], axis=1)
```

### Dealing with missing values

```python
data.isnull().sum()
```

### Visualizing missing values in heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.isnull(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
plt.show()
```

Price and EngineV contain missing values.

### Dropping the missing values

```python
data = data.dropna(axis=0)
```

```python
data.describe(include='all')
```

-  std of price variable is too high
-  huge difference between min and max value shows the presence of outliers

### Dealing with outliers
Get the 99th percentile and keep the data below it. Remove data above 99th percentile.

```python
q = data['Price'].quantile(0.99)
data = data[data['Price'] < q]
```

```python
q = data['Mileage'].quantile(0.99)
data_2 = data[data['Mileage'] < q]
```

```python
data_3 = data_2[data_2['EngineV'] < 6.5]
```

```python
sns.displot(data['Year'])
```

In this case we keep data higher than 1st percentile as there are more newer cars.

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year'] > q]
```

```python
data = data_4.reset_index(drop=True)
```

```python
data.describe(include='all')
```

### Price

```python
sns.displot(data['Price'])
```

-  The data distribution shown above violates linearity
-  Log transformation is used when dealing with exponential relationships

```python
log_price = np.log(data['Price'])
data['log_price'] = log_price
data = data.drop(['Price'], axis=1)
```

```python
sns.displot(data['log_price'])
```

### Correlation between the features in the dataset

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="Blues", fmt='.2f')
plt.show()
```

### Create dummy variables

```python
data_with_dummies = pd.get_dummies(data, drop_first=True)
```

```python
data_with_dummies.head()
```

### Rearranging the data columns

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price','Mileage', 'EngineV', 'Year', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data = data_with_dummies[cols]
data.head()
```

## Linear Regression model

### Declare the inputs and the target

```python
target = data['log_price']
inputs = data.drop(['log_price'], axis=1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, target, test_size=0.2, random_state=365)
```

###  Create the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

### Plotting the predicted values and targets

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets', size=18)
plt.ylabel('Predictions', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

Plot the residuals. From the regression assumptions, the errors/residuals must be normally distributed with the mean of 0.

```python
sns.displot(y_train - y_hat)
plt.title('Residuals PDF', size=18)
```

The distribution is more skewed to the left. There are certain observations for which (y_train - y_hat) is much lower than the mean. This means a much higher price is predicted than is observed. This can be improved to make the model more effective.

```python
reg.score(x_train, y_train)
```

```python
res = reg.score(x_train, y_train)
res_percent = res*100
result = np.round(res_percent,3)
print(f"The model explains {result}% variability of the data.")
```

### Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test)
plt.xlabel('Targets', size=18)
plt.ylabel('Predictions', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

The data points at the lower and upper ends are dispersed.So, the model is not so good at predicting prices that are too low or too high. 

### Convert predictions into prices from log prices

```python
df = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
```

```python
df['Target'] = np.exp(y_test)
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df['Target'] = np.exp(y_test)
```

```python
df['Residual'] = df['Target'] - df['Prediction']
```

```python
df['Difference%'] = np.absolute(df['Residual']/df['Target']*100)
```

```python
pd.options.display.max_rows = 999
pd.set_option('display.float_format', lambda x: '%.2f' %x)
df.sort_values(by=['Difference%'])
```
